# CircleSectorPixRegion Illustration

```python
import numpy as np
from astropy import units as u
from regions import PixCoord, CircleSectorPixelRegion
import matplotlib.pyplot as plt

def plot_quadrants(region):
    for angle in [0, 90, 180, 270] * u.deg:
        region_rot = region.rotate(center=reg.center, angle=angle)
        plot_region(region_rot)

        
def plot_region(region, nx=100, ny=100):
    fig, ax = plt.subplots(1, 1)

    y, x = np.arange(ny), np.arange(nx)
    y, x = np.meshgrid(y, x)

    coords = PixCoord(x=x, y=y)

    mask = region.contains(coords)

    ax.pcolormesh(x, y, mask)

    ax.set_xlim(0, nx)
    ax.set_ylim(0, ny)
    ax.set_aspect('equal')
    patch = region.plot(ax=ax, facecolor='none', edgecolor='red', lw=1,
                     label='Circle')
    bbox = region.bounding_box.plot(ax=ax, color="w", lw=1, label="Bbox")
    ax.legend(handles=(patch, bbox), loc='upper left')
    
```

```python
reg = CircleSectorPixelRegion(PixCoord(x=50, y=50), radius=33, angle_start=34 * u.deg, angle_stop=57 * u.deg)

plot_quadrants(reg)
```

```python
reg = CircleSectorPixelRegion(PixCoord(x=50, y=50), radius=33, angle_start=0 * u.deg, angle_stop=120 * u.deg)

plot_quadrants(reg)
```
